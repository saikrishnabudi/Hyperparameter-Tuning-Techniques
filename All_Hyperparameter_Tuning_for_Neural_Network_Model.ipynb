

Here we are Experiment on all the hyperparameter tuning techniques on the    binary classification diabetes dataset for Neural Network model using Keras & TensorFlow libraries and finally and we can able to see the all the scores.

- Batch Size

  The number of samples trained before the model was updating.

- Epoch

  On epoch trains the entire dataset, In the training process it updates the weights based on the gradient.

- Learning Rate

 The weights of the learning rate network is updated during training and it controls the step size and gives better convergence.

- Dropout Rate

  dropout rate means it will drop the units randomly from neural network to prevent overfitting.

- Activation Function

  Activation function are applied output to each neuron to check whether it should be active or not? some common activation function are sigmoid, tanh, relu.

- Kernel Initializer

  It initializes the weights to neural network with proper process it will speed up the convergence & improve performance.

- Neurons

  Neurons means can changes the capacity & complexity of model.

- Activation Layer

  In Neural Network the activation function are applied to activation layers which introduce the non-linearity and learns the complex patterns.

# **Hyperparameter Tuning for Batch Size & Epochs**

```python
from sklearn.model_selection import GridSearchCV   # It is used for to select best hyperparamenters values
from keras.models import Sequential                # It is used to build the Neural Network Model
from keras.layers import Dense                     # Hidden layers
from keras.wrappers.scikit_learn import KerasClassifier  # We have to put this Neural Network model into Keras classifier and we have to run the model
from sklearn.model_selection import KFold, GridSearchCV
from tensorflow.keras.optimizers import Adam   # Adam Optimizer
```

```python
# Function to create the neural network model
def create_model():
    model = Sequential()
    model.add(Dense(12, input_dim=8, kernel_initializer='random_uniform', activation='relu'))
    model.add(Dense(8, kernel_initializer='random_uniform', activation='relu'))
    model.add(Dense(1, kernel_initializer='random_uniform', activation='sigmoid'))

    # Compiling the model
    adam = Adam(learning_rate=0.01)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model
```

```python
# Define the Keras classifier
model = KerasClassifier(build_fn=create_model, verbose=0)

# Define the grid search parameters
batch_size = [10, 20, 40]
epochs = [10, 50, 100]

# Making the dictonary for Grid Search Parameters
param_grid = dict(batch_size=batch_size, epochs=epochs)

# Build and fit the grid search CV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=KFold(), verbose=10)
grid_result = grid.fit(X_standardized, y)
```

```python
# Summarize the results
print("Best:{},using{}".format(grid_result.best_score_,grid_result.best_params_))
means=grid_result.cv_results_["mean_test_score"]
stds=grid_result.cv_results_["std_test_score"]
params=grid_result.cv_results_["params"]
for mean,stdev,param in zip(means,stds,params):
  print("{},{} with:{}".format(mean,stdev,param))
```

- From the above results we can able to see that from what Batch Size -20 & Which Epoch - 10 is giving the Highest accuracy " Best:0.7657329797744751 "

# **Hyperparameter Tuning for Learning Rate & Dropout Rate**

- Dropout - We can drop the neurons from the hidden layers which can prevent the model from overfitting.

```python
from keras.layers import Dropout

# Function to create and define the neural network model
def create_model(learning_rate,dropout_rate):
    model = Sequential()
    model.add(Dense(8, input_dim=8, kernel_initializer="normal", activation="relu"))
    model.add(Dropout(dropout_rate))
    model.add(Dense(4,input_dim=8,kernel_initializer="normal", activation="relu"))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='sigmoid'))

    # Compiling the model
    adam = Adam(learning_rate=learning_rate)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

# Create the model
model = KerasClassifier(build_fn=create_model, verbose=0, batch_size=40, epochs=10)

# Define the grid search parameters
learning_rate= [0.001,0.01,0.1]
dropout_rate = [0.0,0.1,0.2]

# Make a dictionary of grid search parameters
param_grids = dict(learning_rate=learning_rate,dropout_rate=dropout_rate)

# Build and fit the grid search CV
grid = GridSearchCV(estimator=model, param_grid=param_grids, cv=KFold(), verbose=10)
grid_result = grid.fit(X_standardized, y)
```

```python
# Summarize the results
print("Best:{},using{}".format(grid_result.best_score_,grid_result.best_params_))
means=grid_result.cv_results_["mean_test_score"]
stds=grid_result.cv_results_["std_test_score"]
params=grid_result.cv_results_["params"]
for mean,stdev,param in zip(means,stds,params):
  print("{},{} with:{}".format(mean,stdev,param))
```

- From the above results we can able to understand that from the Drop Out Rate (0.2) & Learning Rate (0.01) we got the best accuracy score is " Best:0.764417290687561 "

# **Hyperparameter Tuning for Activation Function & Kernel Intializer**

```python
# Function to create and define the neural network model
def create_model(activation_function,init):
    model = Sequential()
    model.add(Dense(8, input_dim=8, kernel_initializer=init, activation=activation_function))
    model.add(Dropout(0.1))
    model.add(Dense(4,input_dim=8,kernel_initializer=init, activation=activation_function))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))

    # Compiling the model
    adam = Adam(learning_rate=0.01)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

# Create the model
model = KerasClassifier(build_fn=create_model, verbose=0, batch_size=40, epochs=10)

# Define the grid search parameters
activation_function= ["softmax","relu","tanh","linear"]
init = ["uniform","normal","zero"]

# Make a dictionary of grid search parameters
param_grids = dict(activation_function=activation_function,init=init)

# Build and fit the grid search CV
grid = GridSearchCV(estimator=model, param_grid=param_grids, cv=KFold(), verbose=10)
grid_result = grid.fit(X_standardized,y)
```

```python
# Summarize the results
print("Best:{},using{}".format(grid_result.best_score_,grid_result.best_params_))
means=grid_result.cv_results_["mean_test_score"]
stds=grid_result.cv_results_["std_test_score"]
params=grid_result.cv_results_["params"]
for mean,stdev,param in zip(means,stds,params):
  print("{},{} with:{}".format(mean,stdev,param))
```

# **Tuning of Hyperparameter - Number of Neurons in Activation Layer **

```python
# Function to create the neural network model
def create_model(neuron1, neuron2):
    model = Sequential()
    model.add(Dense(neuron1, input_dim=8, kernel_initializer="uniform", activation="tanh"))
    model.add(Dropout(0.2))
    model.add(Dense(neuron2, input_dim=neuron1, kernel_initializer="uniform", activation="tanh"))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))

    # Compiling the model
    adam = Adam(learning_rate=0.001)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

# Create the model
model = KerasClassifier(build_fn=create_model, verbose=0, batch_size=40, epochs=10)

# Define the Grid Search Parameters
neuron1 = [4, 8, 16]
neuron2 = [2, 4, 8]

# Make a dictionary of grid search parameters
param_grids = dict(neuron1=neuron1, neuron2=neuron2)

# Build and fit the grid search CV
grid = GridSearchCV(estimator=model, param_grid=param_grids, cv=KFold(), verbose=10)
grid_result = grid.fit(X_standardized, y)
```

```python
# Summarize the results
print("Best : {}, using{}".format(grid_result.best_score_,grid_result.best_params_))
means=grid_result.cv_results_["mean_test_score"]
stds=grid_result.cv_results_["std_test_score"]
params=grid_result.cv_results_["params"]
for mean,stdev,param in zip(means,stds,params):
  print("{},{} with:{}".format(mean,stdev,param))
```

- From the above results we can able to see that the best accuracy score is given by neuron1 - 8 & neuron2 - 4 which is " Best : 0.7644172787666321 "

```python
def create_model(learning_rate,dropout_rate,activation_function,init,neuron1,neuron2):
    model = Sequential()
    model.add(Dense(neuron1,input_dim = 8,kernel_initializer = init,activation = activation_function))
    model.add(Dropout(dropout_rate))
    model.add(Dense(neuron2,input_dim = neuron1,kernel_initializer = init,activation = activation_function))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1,activation = 'sigmoid'))

    adam = Adam(learning_rate = learning_rate)
    model.compile(loss = 'binary_crossentropy',optimizer = adam,metrics = ['accuracy'])
    return model

# Create the model

model = KerasClassifier(build_fn = create_model,verbose = 0)

# Define the grid search parameters

batch_size = [10,20,40]
epochs = [10,50,100]
learning_rate = [0.001,0.01,0.1]
dropout_rate = [0.0,0.1,0.2]
activation_function = ['softmax','relu','tanh','linear']
init = ['uniform','normal','zero']
neuron1 = [4,8,16]
neuron2 = [2,4,8]

# Make a dictionary of the grid search parameters

param_grids = dict(batch_size = batch_size,epochs = epochs,learning_rate = learning_rate,dropout_rate = dropout_rate,
                   activation_function = activation_function,init = init,neuron1 = neuron1,neuron2 = neuron2)

# Build and fit the GridSearchCV

grid = GridSearchCV(estimator = model,param_grid = param_grids,cv = KFold(),verbose = 10)
grid_result = grid.fit(X_standardized,y)

# Summarize the results
print('Best : {}, using {}'.format(grid_result.best_score_,grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
  print('{},{} with: {}'.format(mean, stdev, param))
```
